## nwt_QRUNOFF
- Reads in h1 files from moist meadow and 
- Writes .csv to be added to wet meadow precipitation in `prepare_forcings_for_clm.R`

```python
import xarray as xr
import pandas as pd
```

```python
years = range(2008,2020)
case = 'clm50bgc_NWT_mm_newPHS_lowSLA'
dirIN = '/glade/scratch/wwieder/archive/'+case+'/lnd/hist/'
CLMfile = [dirIN+case +'.clm2.h1.'+str(years[i]) +'-01-01-00000.nc' for i in range(len(years)) ]  
CLMfile[0]
```

```python
dsCLM = xr.open_mfdataset(CLMfile, decode_times=True, combine='by_coords') 
dsCLM = dsCLM.isel(lndgrid=0)
dsCLM = dsCLM.shift(time=-1)
dsCLM
```

```python
csvOUT = dirIN+'QRUNOFF_'+case+'.csv'
df = dsCLM.QRUNOFF.to_dataframe(name='QRUNOFF')
print(df.head())
df.to_csv(csvOUT)
print(' --- wrote '+csvOUT+' --- ')
```
